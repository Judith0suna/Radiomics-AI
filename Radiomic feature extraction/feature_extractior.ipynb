```python
from radiomics import featureextractor
import os
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
def radiomic_features_extraction(dataDir, params):
    df = pd.DataFrame()

    count = 0
    for path in os.listdir(dataDir):
        if os.path.isdir(os.path.join(dataDir, path)):
            count += 1

    for i in range(1, count + 1):
        imagePath = os.path.join(dataDir, str(i).zfill(2) + "/" +"Image.nii")
        labelPath = os.path.join(dataDir, str(i).zfill(2) + "/" +"Mask.nii")

        print(imagePath)
        print(labelPath)

        extractor = featureextractor.RadiomicsFeatureExtractor(params)
        extractor.enableAllFeatures()

        try:
            result = extractor.execute(imagePath, labelPath)

            df = df._append(result, ignore_index=True)
        except Exception as e:
            print(f"Error processing case {i}: {e}")

    for i in df.columns.values:
        if "diagnostics" in i:
            df.drop([i], axis=1, inplace=True)

    return df
```
